```python
LambdaSet = np.linspace(-0.045, 0.09, num=2000)
```

```python
SolutionSet = LineUpGen.returnSetOptLineUps(LambdaSet, Known=Know, Pred=Prediction)
```

```python
ePoints = []
Var = []
for sol in SolutionSet:
    if sol[2] > 0:
        ePoints.append(sol[1])
        Var.append(sol[2])
```

```python
#fig, ax = plt.subplots()
#ax.plot(LambdaSet, np.array(ePoints))
#ax.plot(LambdaSet, np.array(Var)/20)
```

```python
Points = [ 90, 100, 105, 110, 115, 120, 125,130, 135, 150, 170, 200, 240, 270]
MaxProb = []
iMaxProb = []
Solutions = []
for j in range(len(Points)):
    Probs = []
    for i in range(len(ePoints)):
        #Probs.append(1-stats.laplace.cdf(Points[j], loc=ePoints[i], scale=np.sqrt(Var[i]/2)))
        Probs.append(1-stats.norm.cdf(Points[j], loc=ePoints[i], scale=np.sqrt(Var[i])))
    iMaxProb.append(np.argmax(Probs))
    MaxProb.append(Probs[np.argmax(Probs)])
iMaxProbU = list(set(iMaxProb))

for i in range(len(iMaxProbU)):
    Solutions.append(SolutionSet[iMaxProbU[i]][0])
```
