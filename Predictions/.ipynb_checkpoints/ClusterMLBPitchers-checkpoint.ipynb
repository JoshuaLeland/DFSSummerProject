```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
LowerGameBound = 100
```

```python
def cleanGL(x):
    try:
        out = float(x)
    except:
        try:
            out = float(x.strip('%'))
        except:
            out = 0
    return out
```

```python
def MakeStdRowPitcher(GameLog):
    try:
        GameLog=GameLog.drop(['GS_x','GSv2'], axis=1)
    except:
        print("Didn't drop GS_x and GSv2")
    cleanRowNorm = [x for x in map(cleanGL,GameLog.values[0,3:])]
    G = cleanRowNorm[4]
    IP = cleanRowNorm[10]
    cleanRowNorm2 = cleanRowNorm[:4]+cleanRowNorm[5:10]+cleanRowNorm[11:]
    cleanRowNorm3 = [x/G for x in cleanRowNorm2[:3]] +[cleanRowNorm2[3]]+ [x/G for x in cleanRowNorm2[4:9]]+[x/IP for x in cleanRowNorm2[9:20]]+[cleanRowNorm2[20]/G]+[x for x in cleanRowNorm2[21:]]
    return cleanRowNorm3
```

```python
def MakeHandRowPitcher(SplitLogHand):
    IPL = SplitLogHand.values[0,2]
    cleanRowSplitvL = [x/IPL for x in map(cleanGL, SplitLogHand.values[0,4:11])]+[x for x in map(cleanGL, SplitLogHand.values[0,11:26])]+[x for x in map(cleanGL, SplitLogHand.values[0,27:])]
    IPR = SplitLogHand.values[1,2]
    cleanRowSplitvR = [x/IPR for x in map(cleanGL, SplitLogHand.values[1,4:11])]+[x for x in map(cleanGL, SplitLogHand.values[1,11:26])]+[x for x in map(cleanGL, SplitLogHand.values[1,27:])]
    return cleanRowSplitvL + cleanRowSplitvR
```

```python
CurrentPitcherRoster = pd.read_csv('../Data/DataBaseCrawler/MLBPitcherRoster2018.csv', index_col=0)
CurrentPitcherRoster = CurrentPitcherRoster[['Name','Team','ID']]
CurrentPitcherRoster.drop_duplicates(subset=['Name','ID'], keep='first', inplace=True)
CurrentPitcherRoster.shape
```

```python
AggregatedPitcherDataFrame=pd.read_csv('../Data/DataBaseCrawler/GameLogs/InactivePitcherList.csv', index_col=0)
AggregatedPitcherDataFrame.drop_duplicates(subset=['ID'], inplace=True)
AggregatedPitcherDataFrame = AggregatedPitcherDataFrame[~AggregatedPitcherDataFrame['ID'].isin(CurrentPitcherRoster['ID'])]
```

```python
#Get the current players
NamesPredictee = []
NamesPredictor = []
CantPredict = []
IDPredictee = []
IDPredictor = []
DataPredictor = []
DataPredictee = []
ActivePredictor = []
ActivePredictee = []
for row in CurrentPitcherRoster.values:
    #Grab the row
    ID = row[2]

    #Grab the GameLog
    GameLog = pd.read_csv('../Data/DataBaseCrawler/GameLogs/Pitchers/'+str(ID)+'GL.csv',index_col=0)

    #Grab the splitLogs
    SplitLogHand = pd.read_csv('../Data/DataBaseCrawler/GameLogs/Pitchers/'+str(ID)+'HandSL.csv', index_col=0)

    #Get the fantasy points values.
    Var = np.var(GameLog['DKfp'][1:])
    mean = np.mean(GameLog['DKfp'][1:])

    #Build the row
    if GameLog.values[0,7] > LowerGameBound and GameLog.values[0,3]/GameLog.values[0,7]>.5:
        NamesPredictor.append(row[0])
        IDPredictor.append(ID)
        ActivePredictor.append('Y')
        RowSTD = MakeStdRowPitcher(GameLog) +[Var]
        RowSLHand = MakeHandRowPitcher(SplitLogHand)
        DataPredictor.append(RowSTD+RowSLHand)
    else:
        if SplitLogHand.values[0,2] > 0 and SplitLogHand.values[1,2] > 0:
            NamesPredictee.append(row[0])
            IDPredictee.append(ID)
            ActivePredictee.append('Y')
            RowSTD = MakeStdRowPitcher(GameLog) +[Var]
            RowSLHand = MakeHandRowPitcher(SplitLogHand)
            DataPredictee.append(RowSTD+RowSLHand)
        else:
            CantPredict.append([row[0], ID])
            
    
PlayersFeatures = np.array(DataPredictor)
PlayersNames = NamesPredictor
IDList = IDPredictor
Active = ActivePredictor
```

```python
#BadIDs=[1462,8127,672,102,189,396,459,1303,60,642,996,920,1273,39,483,794,75,76,1292,1485,1662,1426,929,940,1282,8217,672,742,862,1706,265,1680,137,842,72,1956,758,250,625,1060,1289,191,1453,667,4831,2743,1276,1476]
BadIDs = [72,8217,250,625,1706,1680]
```

```python
InActiveFeatures=[]
InActiveIDs=[]
InActiveNames=[]
InActiveActive=[]
```

```python
#Get the old players
for row in AggregatedPitcherDataFrame.values:
    #row = AggregatedBatterDataFrame
    #Grab the row
    ID = row[1]

    #Grab the GameLog
    GameLog = pd.read_csv('../Data/DataBaseCrawler/GameLogs/InactivePitchers/'+str(ID)+'GL.csv', index_col=0)

    #Grab the splitLogs
    SplitLogHand = pd.read_csv('../Data/DataBaseCrawler/GameLogs/InactivePitchers/'+str(ID)+'HandSL.csv',index_col=0)

    #Get the fantasy points values.
    mean = np.mean(GameLog['DKfp'][1:])
    Var = np.var(GameLog['DKfp'][1:])

    #Build the row - not in BadIDs
    if GameLog.values[0,7] > LowerGameBound and ID not in BadIDs and GameLog.values[0,3]/GameLog.values[0,7]>0.8:
        if SplitLogHand.values[0,2] > 0 and SplitLogHand.values[1,2] > 0:
            IDList.append(ID)
            PlayersNames.append(row[0])
            Active.append('N')
            RowSTD = MakeStdRowPitcher(GameLog)+[Var]
            RowSLHand = MakeHandRowPitcher(SplitLogHand)
            PlayersFeatures=np.concatenate((PlayersFeatures,np.array([RowSTD+RowSLHand])), axis=0)
    else:
        if SplitLogHand.values[0,2] > 0 and SplitLogHand.values[1,2] > 0 and GameLog.values[0,3]/GameLog.values[0,7]>0:
            InActiveIDs.append(ID)
            InActiveNames.append(row[0])
            InActiveActive.append('N')
            RowSTD = MakeStdRowPitcher(GameLog)+[Var]
            RowSLHand = MakeHandRowPitcher(SplitLogHand)
            InActiveFeatures.append(RowSTD+RowSLHand)
```

```python
scaler = StandardScaler(copy=False)
scaler.fit(PlayersFeatures)
ScaledPlayerFeatures = scaler.transform(PlayersFeatures)
ScaledDataPredictee = scaler.transform(DataPredictee)
ScaledInActiveFeatures = scaler.transform(InActiveFeatures)
ScaledPlayerFeatures.shape
```

```python
#Cluster the players - Try a number of clusters and look at error.
Errors = []
NumClusters = range(0,25)
kmeans = [KMeans(n_clusters=i+1) for i in NumClusters]
for model in kmeans:
    model.fit(ScaledPlayerFeatures)
    Errors.append(model.score(ScaledPlayerFeatures))
    print("K: %d Error: %f " % (model.n_clusters, model.score(ScaledPlayerFeatures)));print(np.bincount(model.labels_))
```

```python
plt.plot(NumClusters,Errors)
plt.ylabel('Score')
plt.xlabel("Number of Clusters")
plt.show()
```

```python
ClusterNum = 8
plt.hist(kmeans[ClusterNum-1].labels_)
plt.title("Kmeans clusters size "+str(ClusterNum))
plt.xlabel("Catagory")
plt.ylabel("Frequency")
plt.show()
```

```python
np.bincount(kmeans[ClusterNum-1].labels_)
```

```python
#[3, 12, 20, 41, 49] = [250,625]
PredictorDF = pd.DataFrame()
PredictorDF['Name'] = PlayersNames
PredictorDF['ID'] = IDList
PredictorDF['Labels'] = kmeans[ClusterNum-1].labels_
PredictorDF['Active'] = Active
BadLabelDF = PredictorDF[PredictorDF['Labels']==1]
BadLabelDF
```

```python
#Predict other Batters.
PredicteeLabels = kmeans[ClusterNum-1].predict(ScaledDataPredictee)
InActiveLabels = kmeans[ClusterNum-1].predict(ScaledInActiveFeatures)
TotalLabels = np.concatenate((kmeans[ClusterNum-1].labels_,PredicteeLabels,InActiveLabels))
TotalPlayersNames = PlayersNames + NamesPredictee + InActiveNames
IDListTotal = IDList + IDPredictee + InActiveIDs
TotalActive = Active + ActivePredictee + InActiveActive
```

```python
TotalDF = pd.DataFrame()
TotalDF['Name'] = TotalPlayersNames
TotalDF['ID'] = IDListTotal
TotalDF['Labels'] = TotalLabels
TotalDF['Active'] = TotalActive
```

```python
#Remove clusters with no active players.
ActiveClusters = []
for label in range(ClusterNum):
    #There is probably a better way to do this, but whatever.
    Players = TotalDF.values[[x[2]==label for x in TotalDF.values ]]
    if 'Y' in Players[:,3]:
        ActiveClusters.append(label)
print([item for item in range(ClusterNum) if item not in ActiveClusters])
```

```python
#Remove players with inactive clusters
TotalDF = TotalDF[TotalDF['Labels'].isin(ActiveClusters)]
#Look at final bin count.
np.bincount(TotalDF['Labels'])
```
