```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import pandas as pd
```

```python
def cleanRow(x):
    try:
        return float(x.strip('%'))
    except:
        try:
            return float(x)
        except:
            return x
```

```python
def scrapeTeamPage(Link, year):
    #Scrape STD stats
    TeamData=[]
    soup = BeautifulSoup(requests.get(Link).text, "html.parser")
    Rows = soup.find_all('div','RadGrid RadGrid_FanGraphs')[0].find_all('table')[0].find_all('tr','rgRow')

    for row in Rows:
        Elements = row.find_all('td')
        cleanedRow = [year]+[cleanRow(x.get_text()) for x in Elements[1:]]
        TeamData.append(cleanedRow)
        
    Rows = soup.find_all('div','RadGrid RadGrid_FanGraphs')[0].find_all('table')[0].find_all('tr','rgAltRow')
    for row in Rows:
        Elements = row.find_all('td')
        cleanedRow = [year]+[cleanRow(x.get_text()) for x in Elements[1:]]
        TeamData.append(cleanedRow)
                
    return TeamData
```

```python
#Scrape Batting info
TeamDataSTD=[]
TeamDataADV=[]
for Year in range(1990,2019):
    LinkSTD ='https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=0&type=0&season='+str(Year)+'&month=0&season1='+str(Year)+'&ind=0&team=0,ts&rost=0&age=0'
    LinkADV = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=0&type=1&season='+str(Year)+'&month=0&season1='+str(Year)+'&ind=0&team=0,ts&rost=0&age=0'

    #STD Team Stats
    TeamDataSTD.append(pd.DataFrame(data=scrapeTeamPage(LinkSTD, Year), columns=['Year','Team','G','AB','PA','H','1B','2B','3B','HR','R','RBI','BB','IBB','SO','HBP','SF','SH','GDP','SB','CS','AVG']))

    #Adv Team Stats
    TeamDataADV.append(pd.DataFrame(data=scrapeTeamPage(LinkADV, Year), columns=['Year','Team','PA','BB%','K%','BB/K','AVG','OBP','SLG','OPS','ISO','Spd','BABIP','UBR','wGDP','wSB','wRC','wRAA','wOBA','wRC+']))

#Concat and merge the results.
TeamDataStd = pd.concat(TeamDataSTD)
TeamDataAdv = pd.concat(TeamDataADV)

#Merge
TeamBattingData = pd.merge(TeamDataStd, TeamDataAdv, on=['Year', 'Team', 'PA', 'AVG'])
```

```python
TeamBattingData.to_csv('./TeamStats/TeamBattingStats.csv')
```

```python
#Scrape Pitching info
TeamDataSTD=[]
TeamDataADV=[]
for Year in range(1990,2019):
    LinkSTD ='https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=all&qual=0&type=0&season='+str(Year)+'&month=0&season1='+str(Year)+'&ind=0&team=0,ts&rost=0&age=0'
    LinkADV = 'https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=all&qual=0&type=1&season='+str(Year)+'&month=0&season1='+str(Year)+'&ind=0&team=0,ts&rost=0&age=0&filter=&players=0'

    #STD Team Stats
    TeamDataSTD.append(pd.DataFrame(data=scrapeTeamPage(LinkSTD, Year), columns=['Year','Team','W','L','ERA','G','GS','CG','ShO','SV','HLD','BS','IP','TBF','H','R','ER','HR','BB','IBB','HBP','WP','BK','SO']))

    #Adv Team Stats
    TeamDataADV.append(pd.DataFrame(data=scrapeTeamPage(LinkADV, Year), columns=['Year','Team','K/9','BB/9','K/BB','HR/9','K%','BB%','K-BB%','AVG','WHIP','BABIP','LOB%','ERA-','FIP-','xFIP-','ERA','FIP','E-F','xFIP', 'SIERA']))

#Concat and merge the results.
TeamDataStd = pd.concat(TeamDataSTD)
TeamDataAdv = pd.concat(TeamDataADV)

#Merge
TeamPitchingData = pd.merge(TeamDataStd, TeamDataAdv, on=['Year', 'Team', 'ERA'])
```

```python
TeamPitchingData.to_csv('./TeamStats/TeamPitchingStats.csv')
```
