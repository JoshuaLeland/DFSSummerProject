```python
import pandas as pd
```

```python
#INIT SECTION -- 
DKSalaries = 'DKSalaries/DKSalariesJuly10.csv'
RawFile = pd.read_csv(DKSalaries)
```

```python
def Abv2Name(x):
    return {
        'ARI': 'Diamondbacks',
        'ATL': 'Braves',
        'BAL': 'Orioles',
        'CHC': 'Cubs',
        'CLE': 'Indians',
        'COL': 'Rockies',
        'DET': 'Tigers',
        'MIA': 'Marlins',
        'HOU': 'Astros',
        'LAA': 'Angels',
        'LAD': 'Dodgers',
        'MIL': 'Brewers',
        'MIN': 'Twins',
        'NYM': 'Mets',
        'NYY': 'Yankees',
        'OAK': 'Athletics',
        'PHI': 'Phillies',
        'PIT': 'Pirates',
        'SD': 'Padres',
        'SF': 'Giants',
        'SEA': 'Mariners',
        'STL': 'Cardinals',
        'TEX': 'Rangers',
        'TB': 'Rays',
        'TOR': 'Blue Jays',
        'WAS': 'Nationals',
        'BOS': 'Red Sox',
        'CIN': 'Reds',
        'CWS': 'White Sox',
        'KC': 'Royals',
        'CHW': 'White Sox',
        'KCR' : 'Royals',
        'SDP' : 'Padres',
        'SFG' : 'Giants',
        'TBR' : 'Rays'
    }[x]
```

```python
RawFile['Team'] = RawFile.apply(lambda x: Abv2Name(x['TeamAbbrev']), axis=1)
```

```python
RawFile
```

```python
FormattedSalaries = pd.DataFrame()
FormattedSalaries['Name'] = RawFile['Name']
FormattedSalaries['Team'] = RawFile['Team']
FormattedSalaries['Position'] = RawFile['Position']
FormattedSalaries['Salary'] = RawFile['Salary']
```

```python
FormattedSalaries
```

```python
#Attach Fangraphs predictions
BatterPred = 'FanGraphPred/July10/FanGraphsBatterJuly10.csv'
PitcherPred = 'FanGraphPred/July10/FanGraphsPitcherJuly10.csv'
```

```python
#Load the prediction files.
BatterPredDF = pd.read_csv(BatterPred)
PitcherPredDF = pd.read_csv(PitcherPred)
```

```python
#Drop useless columns
BatterPredDF.drop(columns=['Game','Pos','PA','H','1B','2B','3B','HR','R','RBI','SB','CS','BB','SO','Yahoo','FanDuel', 'playerid'], inplace=True)
PitcherPredDF.drop(columns=['Game','W','IP','TBF','H','1B','2B','3B','HR','BB','SO','Yahoo','FanDuel', 'playerid'], inplace=True)
```

```python
FanGraphPred = pd.concat([BatterPredDF,PitcherPredDF]).rename(columns={'DraftKings':'DKFG'})
```

```python
FanGraphPred.head()
```

```python
#Merge the salaries with original.
FormattedSalaries = FormattedSalaries.merge(FanGraphPred, how='left', left_on=['Name', 'Team'], right_on=['Name', 'Team']).dropna(subset=['DKFG'])
```

```python
FormattedSalaries.head()
```

```python
#Add in the Bat projections.
TheBatBatter = './The-BAT/the-bat-hitters-Aug18.csv'
TheBatPitcher = './The-BAT/the-bat-pitchers-Aug18.csv'
BatBatterPred = pd.read_csv(TheBatBatter)
BatPitcherPred = pd.read_csv(TheBatPitcher)
```

```python
BatBatterPred = BatBatterPred[['NAME', 'TEAM', 'FPTS']]
BatBatterPred['TEAM']=BatBatterPred.apply(lambda x: Abv2Name(x['TEAM']), axis=1)
BatBatterPred.columns = ['Name', 'Team', 'BATPred']
BatBatterPred.head()
```

```python
BatPitcherPred = BatPitcherPred[['NAME', 'TEAM', 'FPTS']]
BatPitcherPred['TEAM'] = BatPitcherPred.apply(lambda x: Abv2Name(x['TEAM']), axis=1)
BatPitcherPred.columns = ['Name', 'Team', 'BATPred']
BatPitcherPred.head()
```
