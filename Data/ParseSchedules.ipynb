```python
import pandas as pd
import numpy as np
```

```python
Years = range(1990, 2018)
```

```python
year = 2015
#for year in Years:

rLog = pd.read_csv("./RetrosheetData/GameLogs/gl"+str(year)+".txt", header=None)


Data = []
Columns=['Date','Time', 'Stadium ID','Umpire ID','Umpire Name','Home Team','Visiting Team','Home Pitcher ID','Home Pitcher Name','Away Pitcher ID','Away Pitcher Name','HBO 1 ID','HBO 1 Name','HBO 1 Pos Key','HBO 2 ID','HBO 2 Name','HBO 2 Pos Key', 'HBO 3 ID','HBO 3 Name','HBO 3 Pos Key','HBO 4 ID','HBO 4 Name','HBO 4 Pos Key','HBO 5 ID','HBO 5 Name','HBO 5 Pos Key','HBO 6 ID','HBO 6 Name','HBO 6 Pos Key','HBO 7 ID','HBO 7 Name','HBO 7 Pos Key','HBO 8 ID','HBO 8 Name','HBO 8 Pos Key','HBO 9 ID','HBO 9 Name','HBO 9 Pos Key','ABO 1 ID','ABO 1 Name','ABO 1 Pos Key','ABO 2 ID','ABO 2 Name','ABO 2 Pos Key','ABO 3 ID','ABO 3 Name','ABO 3 Pos Key','ABO 4 ID','ABO 4 Name','ABO 4 Pos Key','ABO 5 ID','ABO 5 Name','ABO 5 Pos Key','ABO 6 ID','ABO 6 Name','ABO 6 Pos Key','ABO 7 ID','ABO 7 Name','ABO 7 Pos Key','ABO 8 ID','ABO 8 Name','ABO 8 Pos Key','ABO 9 ID','ABO 9 Name','ABO 9 Pos Key']
for row in rLog.values:
    #Date
    Date = str(row[0])[:4]+'-'+str(row[0])[4:6]+'-'+str(row[0])[6:8]

    #Time
    Time = row[12]

    #StadiumID
    StadiumID = row[16]

    #HomePlate Umpire
    UmpireH = row[77:79].tolist()

    #HomeTeam
    Home = row[6]

    #AwayTeam
    Away = row[3]

    #Home Team Starting Pitcher
    PitcherStartH = row[103:105].tolist()

    #Away Team Starting Pitcher
    PitcherStartA = row[101:103].tolist()

    #Home Team Roster - In batting order
    HomeTeamRoster = row[132:159].tolist()

    #Away Team Roster - In batting order
    AwayTeamRoster = row[105:132].tolist()

    OutRow = [Date,Time,StadiumID]+UmpireH+[Home,Away]+PitcherStartH+PitcherStartA+HomeTeamRoster+AwayTeamRoster
    
    Data.append(OutRow)
    
OutDF = pd.DataFrame(data=Data,columns=Columns)
OutDF.to_csv('./Schedules/RSSchedule'+str(year)+'.csv')
```
