```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
r = requests.get('https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=0&type=8&season=2018&month=0&season1=2018&ind=0&team=0,ts&rost=&age=&filter=&players=0')
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
table = soup.find_all("td", "grid_line_regular")
```

```python
teamLinksBat =[]
teamname = []
for i in range(0,30):
	teamLinksBat.append(table[i*17+1].find("a").get('href'))
	teamname.append(table[i*17+1].text)
```

```python
teamname
```

```python
teamLinksPit = []
for links in teamLinksBat:
    teamLinksPit.append(links.replace("bat", "pit"))

teamLinksPit
```

```python
dataIn = {'TeamName': teamname, 'TeamLinksBat' : teamLinksBat, 'TeamLinksPit' : teamLinksPit}
```

```python
dataframe = pd.DataFrame(data=dataIn)
```

```python
dataframe = dataframe[['TeamName', 'TeamLinksBat', 'TeamLinksPit']]
```
