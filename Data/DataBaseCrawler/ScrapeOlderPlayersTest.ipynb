```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import bleach
```

```python
#Get Links to all the players.
Link = 'https://www.fangraphs.com/players.aspx'
soup = BeautifulSoup(requests.get(Link).text, "html.parser")
```

```python
PlayerLinks = soup.find_all('div', 's_name')
```

```python
Links = []
for links in PlayerLinks:
    subLinks = links.find_all('a', href=True)
    for l in subLinks:
        Links.append(l['href'])
```

```python
#Groom the links
Linkc = [link.replace("D'","D%27") for link in Links]
Linkc = [link.replace("O'","O%27") for link in Linkc]
```

```python
Linkc
```

```python
PastPlayers = []
for link in Linkc:
    nLink = 'https://www.fangraphs.com/'+link
    soup = BeautifulSoup(requests.get(nLink).text, "html.parser")
    try:
        PlayerInfo = soup.find_all('td', style='vertical-align: top;')[0].find_all('table')[0].find_all('tr')
    except:
        print("Couldn't scrape the following link: "+link)
    for players in PlayerInfo:
        try:
            Name = players.find_all('a')[0].get_text()
            Link = players.find_all('a')[0]['href']
            ID = Link.split('=')[1].split('&')[0]
            Position = Link.split('=')[2]
            Years = players.find_all('td')[1].get_text().split('-')
            sYear = Years[0].strip()
            eYear = Years[1].strip()
            row = [Name, ID, Position, Link, sYear, eYear]
            PastPlayers.append(row)
        except:
            print("Player could not be added")
```

```python
OldRoster = pd.DataFrame(data=PastPlayers, columns=['Name', 'ID', 'Position', 'Link', 'Start Year', 'End Year'])
```

```python
OldRoster.to_csv("./OlderRosters/MasterRosterList.csv")
```
