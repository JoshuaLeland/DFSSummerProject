```python
Link = 'https://www.fangraphs.com/statsplits.aspx?playerid=15640&position=OF&season=0'
```

```python
rSplit = requests.get(Link)
soupSplit = BeautifulSoup(rSplit.text, "html.parser")
```

```python
#Get the Standard table.
RowSet1=soupSplit.find("table",{"class":"rgMasterTable", "id":"SeasonSplits1_dgSeason1_ctl00"}).find_all("tr",{"class":"rgRow"})
RowSet2=soupSplit.find("table",{"class":"rgMasterTable", "id":"SeasonSplits1_dgSeason1_ctl00"}).find_all("tr",{"class":"rgAltRow"})
```

```python
#Get Pitcher info
Link = 'https://www.fangraphs.com/statsplits.aspx?playerid=4930&position=P&season=0'
```

```python
rSplit = requests.get(Link)
soupSplit = BeautifulSoup(rSplit.text, "html.parser")
```

```python
#Get Standard Split Stats
SplitHandedStd=[]
StdTable=soupSplit.find("table",id="SeasonSplits1_dgSeason1_ctl00").find_all('td', 'grid_line_regular')
for i in range(4):
    Line = []
    for j in range(19):
        try:
            Line.append(float(StdTable[19*(i)+j].get_text().strip("%")))
        except:
            Line.append(StdTable[19*(i)+j].get_text())
    SplitHandedStd.append(Line)
SplitHandedStdDF = pd.DataFrame(data=SplitHandedStd[0:2], columns=['Season','Handedness','IP','ERA','TBF','H','2B','3B','R','ER','HR','BB','IBB','HBP','SO','AVG','OBP','SLG','wOBA'])
SplitHandedStdDF
```

```python
SplitHomeAwayStdDF=pd.DataFrame(data=SplitHandedStd[2:4], columns=['Season','Home/Away','IP','ERA','TBF','H','2B','3B','R','ER','HR','BB','IBB','HBP','SO','AVG','OBP','SLG','wOBA'])
SplitHomeAwayStdDF
```

```python
#Get Advanced Split Stats
SplitHandedAdv=[]
StdTable=soupSplit.find("table",id="SeasonSplits1_dgSeason2_ctl00").find_all('td', 'grid_line_regular')
for i in range(4):
    Line = []
    for j in range(15):
        try:
            Line.append(float(StdTable[15*(i)+j].get_text().strip("%")))
        except:
            Line.append(StdTable[15*(i)+j].get_text())
    SplitHandedAdv.append(Line)
SplitHandedAdvDF=pd.DataFrame(data=SplitHandedAdv[0:2], columns=['Season','Handedness','K/9','BB/9','K/BB','HR/9','K%','BB%','K-BB%','AVG','WHIP','BABIP','LOB%','FIP','xFIP'])
SplitHandedAdvDF
```

```python
SplitHomeAwayAdvDF=pd.DataFrame(data=SplitHandedAdv[2:4], columns=['Season','Home/Away','K/9','BB/9','K/BB','HR/9','K%','BB%','K-BB%','AVG','WHIP','BABIP','LOB%','FIP','xFIP'])
```

```python
#Merge Handedness Pitcher
SplitHandedPitcher = pd.merge(SplitHandedStdDF,SplitHandedAdvDF, on=['Season', 'Handedness', 'AVG'])
SplitHandedPitcher
```

```python
#Merge Home/Away Pitcher
SplitHomeAwayPitcher = pd.merge(SplitHomeAwayStdDF,SplitHomeAwayAdvDF, on=['Season', 'Home/Away', 'AVG'])
SplitHomeAwayPitcher
```
