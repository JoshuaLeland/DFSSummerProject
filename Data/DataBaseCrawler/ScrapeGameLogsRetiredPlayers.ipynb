```python
import pandas as pd
import numpy as np
import GameLogCrawler.ScrapeGameLogs as sc
import GameLogCrawler.ScrapeSplitLogs as sp
import os
```

```python
#Upload current Rosters
CurrentBattingRoster = pd.read_csv('MLBBatterRoster2018.csv', index_col=0)
CurrentPitchingRoster = pd.read_csv('MLBPitcherRoster2018.csv', index_col=0)
```

```python
CurrentBattingRoster['Link'][0]
```

```python
#Aggregate no longer active batters.
MasterPlayerDataFrame = pd.read_csv('./OlderRosters/MasterRosterList.csv', index_col=0)
BatterDataFrame = MasterPlayerDataFrame[MasterPlayerDataFrame['Position']!='P']
BatterDataFrame=BatterDataFrame[~BatterDataFrame['ID'].isin(CurrentBattingRoster['ID'])]
```

```python
BatterDataFrame.values
```

```python
#Aggregate no longer active pitchers.
PitcherDataFrame = MasterPlayerDataFrame[MasterPlayerDataFrame['Position']=='P']
PitcherDataFrame=PitcherDataFrame[~PitcherDataFrame['ID'].isin(CurrentPitchingRoster['ID'])]
```

```python
PitcherDataFrame.head()
```

```python
#Get the Old Batter GameLogs
BatterIDs = []
for row in BatterDataFrame.values:
    #Get ID number. -- This is probably a gross way to do this but idc.
    Link = row[3].replace('statss.aspx','statsd.aspx')+'&gds=&gde=&type=1&season=all'
    ID = row[1]
    Pos = row[2]
    
    try:
        #Scrape the id
        scraper = sc.BatterGameLog(Link)
        scraper.scrapeStdGL()
        scraper.scrapeAdvGL()

        #Get SplitLog
        splitscraper = sp.BatterSplitLog('statsplits.aspx?playerid='+str(ID)+'&position='+Pos+'&season=0')
        splitscraper.scrapeSplitLogs()


        #Save the csv as the id number
        scraper.exportGL(str(ID)+"GL", 'GameLogs/InactiveBatters', 'Both')
        splitscraper.exportGL(str(ID), 'GameLogs/InactiveBatters', 'Both')
        if scraper.CanExport:
            BatterIDs.append(ID)
    except:
        print(str(ID)+" could not have it's gamelog/splitlog scraped.")
```

```python
#Save stored Batters
BatterDataFrame=BatterDataFrame[BatterDataFrame['ID'].isin(BatterIDs)]
BatterDataFrame.to_csv("./GameLogs/InactiveBatterList.csv")
```

```python
PitcherIDs=[]
for row in PitcherDataFrame.values:
    #Get ID number. -- This is probably a gross way to do this but idc.
    Link = row[3].replace('statss.aspx','statsd.aspx')+'&gds=&gde=&type=1&season=all'
    ID = row[1]
    Pos = row[2]

    try:
        #Scrape the id
        scraper = sc.PitcherGameLog(Link)
        scraper.scrapeStdGL()
        scraper.scrapeAdvGL()

        #Split Pitcher Log
        splitscraper = sp.PitcherSplitLog('statsplits.aspx?playerid='+str(ID)+'&position=P&season=0')
        splitscraper.scrapeSplitLogs()

        #Save the csv as the id number
        scraper.exportGL(str(ID)+"GL", 'GameLogs/InactivePitchers', 'Both')
        splitscraper.exportGL(str(ID), 'GameLogs/InactivePitchers', 'Both')
        if scraper.CanExport:
            PitcherIDs.append(ID)
    except:
        print(str(ID)+" could not have it's gamelog/splitlog scraped.")
```

```python
PitcherDataFrame=PitcherDataFrame[PitcherDataFrame['ID'].isin(PitcherIDs)]
PitcherDataFrame.to_csv("./GameLogs/InactivePitcherList.csv")
```
